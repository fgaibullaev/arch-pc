```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x):
    return x**3 - 5*x**2 + 1

# Создаем массив значений x
x = np.linspace(-1, 7, 400)
y = f(x)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^3 - 5x^2 + 1')
plt.axhline(0, color='red', linewidth=0.5, linestyle='--')  # Ось абсцисс
plt.axvline(0, color='red', linewidth=0.5, linestyle='--')  # Ось ординат
plt.title('График функции')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
def g(x):
    return (5*x**2 - 1)**(1/3)  # x = g(x)

# Начальное приближение
x0 = 0.5
tolerance = 1e-5  # Заданная точность
max_iterations = 100  # Максимальное количество итераций

def simple_iteration(x0):
    x_n = x0
    for n in range(max_iterations):
        x_next = g(x_n)
        if abs(x_next - x_n) < tolerance:
            return x_next, n + 1
        x_n = x_next
    return x_n, max_iterations

root_simple, iterations_simple = simple_iteration(x0)
print(f"Корень методом простой итерации: {root_simple}, Итерации: {iterations_simple}")
def f_prime(x):
    return 3*x**2 - 10*x  # Производная функции

def newtons_method(x0):
    x_n = x0
    for n in range(max_iterations):
        f_n = f(x_n)
        f_prime_n = f_prime(x_n)

        if f_prime_n == 0:  # Проверка деления на ноль
            print("Производная равна нулю, метод Ньютона остановлен.")
            return None, n

        x_next = x_n - f_n / f_prime_n
        if abs(x_next - x_n) < tolerance:
            return x_next, n + 1
        x_n = x_next
    return x_n, max_iterations

root_newton, iterations_newton = newtons_method(x0)
print(f"Корень методом Ньютона: {root_newton}, Итерации: {iterations_newton}")
```
